```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
pyber_df = pd.merge(ride_data, city_data, on="city", how="left")

# Display the data table for preview
pyber_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_type = pyber_df[pyber_df["type"] == "Urban"].groupby([pyber_df["city"]])
rural_type = pyber_df[pyber_df["type"] == "Rural"].groupby([pyber_df["city"]])
suburban_type = pyber_df[pyber_df["type"] == "Suburban"].groupby([pyber_df["city"]])

# Build the scatter plots for each city types

# Urban
id_urban = urban_type["ride_id"].count()
f_urban = urban_type["fare"].mean()
c_urban = urban_type["driver_count"].mean()

# Rural
id_rural = rural_type["ride_id"].count()
f_rural = rural_type["fare"].mean()
c_rural = rural_type["driver_count"].mean()

# Suburban
id_suburban = suburban_type["ride_id"].count()
f_suburban = suburban_type["fare"].mean()
c_suburban = suburban_type["driver_count"].mean()

# Scattorplots
plt.scatter(id_urban, f_urban, label = "Urban", s=c_urban * 10, color=["coral"], edgecolor="black", alpha = 0.7, marker="o")
plt.scatter(id_rural, f_rural, label = "Rural", s=c_rural * 10, color=["gold"], edgecolor="black", alpha = 0.7, marker="o")
plt.scatter(id_suburban, f_suburban, label = "Suburban", s=c_suburban  * 10, color=["lightskyblue"], edgecolor="black", alpha = 0.7, marker="o")
plt.grid()

# Incorporate the other graph properties
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")

# Create a legend
legend = plt.legend(fontsize = 8.5, title= "City Types", loc="upper right")
legend.legendHandles[0]._sizes = [50]
legend.legendHandles[1]._sizes = [50]
legend.legendHandles[2]._sizes = [50]

# Incorporate a text label regarding circle size
plt.text(42,35,"Note: \nCircle size correlates with driver count per city.")

# Save Figure
plt.savefig("../Images/PyberRideScatterplot.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
pyber_type_df = pyber_df.groupby(['type'])

fare_sum = pyber_type_df['fare'].sum()

# Build Pie Chart
plt.pie(fare_sum, explode=(0, 0, 0.15), labels=["Rural","Suburban","Urban"], colors=["gold", "lightskyblue", "coral"], autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("../Images/TotalFaresPieChart.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rides_count = pyber_type_df['ride_id'].count()

# Build Pie Chart
plt.pie(rides_count, explode=(0, 0, 0.15), labels=["Rural","Suburban","Urban"], colors=["gold", "lightskyblue", "coral"], autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("../Images/TotalRidesPieChart.png")
```

```python
# Show Figure
plt.show()
```

```python
# Calculate Driver Percents
city_type_df = city_data.groupby(['type'])
drivers_sum = city_type_df['driver_count'].sum()

# Build Pie Charts
plt.pie(drivers_sum, explode=(0, 0, 0.15), labels=["Rural","Suburban","Urban"], colors=["gold", "lightskyblue", "coral"], autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("../Images/TotalDriversPieChart.png")
```
